```python
!pip -q install google-genai gradio

import os
from google import genai
import gradio as gr

client=genai.Client(api_key=os.environ.get("GEMINI_API_KEY"))

LANGUAGES = [
    "None (keep original)", "Hindi", "Telugu", "Tamil", "Spanish",
    "French", "German", "Arabic", "Chinese (Simplified)", "Japanese"
]

def generate_blog(topic, audience, tone, words, language):
    prompt = f"""
    Write a {words}-word blog.

    Topic: {topic}
    Audience: {audience}
    Tone: {tone}

    Include:
    -Introduction
    -Fun Facts
    -Advantages
    -Conclusion
    """

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )
    original_text = response.text

    # If a translation language is selected, translate the generated blog
    if language and language != "None (keep original)":
        translate_prompt = f"""
        Translate the following blog post into {language}.
        Keep the formatting (headings, bullet points) intact.
        Do not add any commentary — only return the translated blog.

        Blog:
        {original_text}
        """
        translated_response = client.models.generate_content(
            model="gemini-2.5-flash",
            contents=translate_prompt
        )
        translated_text = translated_response.text
    else:
        translated_text = "*(No translation requested)*"

    return original_text, translated_text


demo = gr.Interface(
    fn=generate_blog,
    inputs=[
        gr.Textbox(label="Topic"),
        gr.Textbox(label="Audience"),
        gr.Dropdown(
            ["Professional", "Casual", "Funny", "Technical", "Inspirational"],
            label="Tone"
        ),
        gr.Slider(200, 1000, value=500, label="Word Count"),
        gr.Dropdown(LANGUAGES, value="None (keep original)", label="Translate To")
    ],
    outputs=[
        gr.Markdown(label="Original Blog (English)"),
        gr.Markdown(label="Translated Blog")
    ],
    title="AI Blog Generator with Multi-Language Translation (Gemini)"
)

demo.launch()
```
